```python
#step 1-Import Libraries
import pandas as pd
import sqlite3
```

```python
#step 2-Create Relational Database
conn=sqlite3.connect("customer.db")
data={
"id":[1,2,3,4],
"name":["Ali","Sara","John","David"],
"age":[25,30,28,35],
"city":["Delhi","Mumbai","London","Dubai"]
}
df=pd.DataFrame(data)
df.to_sql("customers",conn,index=False,if_exists="replace")
print("Database table created")
```

```python
#step 3-Read Data from Database
query = "SELECT * FROM customers"
db_data=pd.read_sql(query,conn)
db_data
```

```python
#step 4-Simulate Import to Hadoop(HDFS)
db_data.to_csv("hdfs_customer_data.csv",index=False)
print("Data imported into simulated HDFS")
```
